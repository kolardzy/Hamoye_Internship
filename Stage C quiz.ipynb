**Question 2**

```python
con1 = Africa_fs['Year']==2004
con2 = Africa_fs['Year']==2005
con3 = Africa_fs['Year']==2006
new_data1 = Africa_fs.loc[con1]
new_data2 = Africa_fs.loc[con2]
new_data3 = Africa_fs.loc[con3]
```

```python
merged = pd.merge(new_data1, new_data2, how = 'outer')
merged
```

```python
merged2 = pd.merge( merged, new_data3, how ='outer')
merged2
```

```python
lowest = merged2.groupby('Country')['Value'].sum().sort_values(ascending = True)
lowest
```

The answer is Zambia, Ethiopia and Central Africa Republic.
