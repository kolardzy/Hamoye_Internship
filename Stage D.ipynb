**Gross domestic product (GDP)** 

As defined on WIkipedia is a monetary measure of the market value of all the final goods and services produced in a specific time period.

The OECD defines GDP as "an aggregate measure of production equal to the sum of the gross values added of all resident and institutional units engaged in production and services (plus any taxes, and minus any subsidies, on products not included in the value of their outputs)". An IMF publication states that, "GDP measures the monetary value of final goods and services—that are bought by the final user—produced in a country in a given period of time (say a quarter or a year)."

Total GDP can also be broken down into the contribution of each industry or sector of the economy. The ratio of GDP to the total population of the region is the per capita GDP and the same is called Mean Standard of Living. GDP is considered the "world's most powerful statistical indicator of national development and progress".

```python
#importing required libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
sns.set(style = 'whitegrid')
plt.rc('figure', figsize=(20,10))
```

```python
import pandas as pd
url = "https://en.wikipedia.org/wiki/List_of_sovereign_states_in_Europe_by_GDP_(nominal)" #data sourced from html table on wikipedia

dfs = pd.read_html(url, header = 0)
df = dfs[1]
df
```

```python
df = df.drop('Unnamed: 0', axis = 1 ) #dropping irrelevant column
```

```python
#renaming a particular column
new_name = {'2019[1]': '2019'}
df = df.rename(columns = new_name)
```

```python
#renaming individual countries
df.iat[47,0] = 'Monaco'
df.iat[48,0] = 'Andorra'
df.iat[49,0] = 'Liechtenstein'
df
```

```python
df_plot = df.drop(['2010','2011','2012','2013','2014','2015', '2016','2017','2018'], axis =1) #we are only using data in the '2019' column.
df_plot
```

```python
df_plots = df_plot[0:19] #Top 19 countries.  
```

```python
import plotly.express as px
df = df_plots
fig = px.bar(df, x="Country", y="2019")
fig.update_layout(uniformtext_minsize=8,title_text='<b>Top</b> 19 European Countries by GDP(Billions $) for 2019',
                 xaxis_title="Country",
                  yaxis_title="Value")
fig.show('svg')
```

As we can see from the column chart, **Germany** was the best performing country for the year 2019 by a considerable margin. United Kingdom and France were close with little difference. On the otherend, Romania was the lowest out of the 19 countries.

**Nominal GDP per capita**. 

Put simply, this is the "average" per-person output of the economy in the prices of the current year.
Per capita gross domestic product (GDP) is a metric that breaks down a country's economic output per person and is calculated by dividing the GDP of a country by its population.
It is a global measure for gauging the prosperity of nations and is used by economists, along with GDP, to analyze the prosperity of a country based on its economic growth.
Small, rich countries and more developed industrial countries tend to have the highest per capita GDP.
Per capita GDP is the most universal because its components are regularly tracked on a global scale, providing for ease of calculation and usage.
At its most basic interpretation. per capita GDP shows how much economic production value can be attributed to each individual citizen. Alternatively, this translates to a measure of national wealth since GDP market value per person also readily serves as a prosperity measure.

```python
df_plots3 = df_plot3[0:19]
```

Interesting to note that **Nigeria** had the highest GDP in Africa for the year 2019, but it is still termed the "Poverty Capital" of the world. This points to a fact that African countries are generally not doing well in terms of economic growth. 

South Africa is the second top country but there is a considerable difference from the Nigerian value.
